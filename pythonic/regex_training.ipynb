```python
import re
```

```python
# [] указываются варианты на выбор. Подойдет любой из них. В том числе можно указывать символы через дефис
# . ^ $ * + { } ( ) [ ] \ | - метасимволы
# \d ~ [0-9]
# \D ~ [^0-9]
# \s ~ [ \t\n\r\f\v] пробельные символы
# \S ~ [^ \t\n\r\f\v]
# \w ~ [a-zA-Z0-9_] цифры, буквы и символ _
# \W ~ [^a-zA-Z0-9_]
#символ точка используется, если нужно найти любой символ в тексте. Например, 'a.c' выдаст любые комбинации, где между a c есть что угодно
# * используется после символа, чтобы найти его, если он повторяется в тексте сколько угодно раз
# + аналогично, но в этом случае не будут найдены выражения, в которых 0 вхождений искомого символа
# ? аналогично, но только для 0 или 1 вхождения искомого символа
# {} для поиска конкретного количества вхождений искомого символа в выражение. {2} - 2, {2,4} - от 2 до 4, то есть и 3 тоже найдется
```

```python
print(re.match)
print(re.search)
print(re.findall)
print(re.sub)
```

```python
pattern = r'a[a-zA-Z]c'
string = 'aЫc, aac, acc, aZc, aqc, a c, a.c, a^c'
match_object = re.match(pattern, string) # inly if string starts with pattern
print(match_object)
```

```python
pattern = r'a[^a-zA-Z]c'
string = 'aЫc, aac, acc, aZc, aqc, a c, a.c, a^c'
match_object = re.match(pattern, string)
print(match_object)
```

```python
search_obj = re.search(pattern, string)
print(search_obj)
```

```python
all_inc = re.findall(pattern, string) # list of all matching patterns
print(all_inc)
```

```python
fixed = re.sub(pattern, 'a', string) #find and substitute pattern in a string
print(fixed)
```

```python
meta_found = re.findall(r'a[\w.^]c', string) # any symbols a-zA-Z0-9_, and '.', and '^' between a and c
meta_found
```

```python
st = 'a/b, a//b, a///b, a////b'
pat = r'a/{1,3}b'
re.findall(pat, st)
```

```python
st = 'ab, abbbbb, abcdbb, adsb'
pat = r'a[bcd]*b'
print(re.findall(pat, st))
print('*'*30)
print('The difference between * and + is that + requires 1 or more matches \
and * can have no matches')
print('*'*30)
pat = r'a[bcd]+b'
print(re.findall(pat, st))
```
